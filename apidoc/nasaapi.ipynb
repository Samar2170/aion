```python
import requests
import dotenv
import os
```

```python
nak=os.getenv("NASA_API_KEY")
```

```python
# Astronomy picture of the day
url="https://api.nasa.gov/planetary/apod"
resp=requests.get(
    url,
    params={
        "api_key":nak
    }

)
```

```python
resp
```

```python
data=resp.json()
```

```python
# Sateliite image of something
lat,long=28.517160813992056, 76.99989929221555
url="https://api.nasa.gov/planetary/earth/assets"

resp=requests.get(
    url,
    params={
        "api_key":nak,
        "lat":lat,
        "lon":long,
        "dim":"0.25"
    }
)
data=resp.json()
```

```python
data
```

```python
# search

url="https://images-api.nasa.gov/search"

resp=requests.get(
    url,
    params={
        "q":"voyager",
        "description":"captured by voyager"
    }
)
data=resp.json()
data
```
